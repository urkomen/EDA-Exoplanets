### 1. Mediante la API

Recuperamos los datos de la [API de la NASA](https://exoplanetarchive.ipac.caltech.edu/docs/program_interfaces.html)  

El dataset es de la tabla Planetary Systems Composite Data

```python
import requests
import pandas as pd
```

Generamos la url y descargamos los datos

```python
tabla = 'query=select+*+from+pscomppars' # Tabla PSCompPars
tipo = 'format=json'
url_nasa = f'https://exoplanetarchive.ipac.caltech.edu/TAP/sync'

url = url_nasa + '?' + tabla + '&' + tipo
print(url)
response = requests.get(url)

if response.status_code == 200:  # Código 200 indica una respuesta exitosa.
    exoplanets_json = response.json()  # Analizar la respuesta JSON.
    print('Respuesta OK -', response.status_code)
else:
    print("Error en la solicitud: ", response.status_code)
```

Guardamos el json descargado en un dataframe y ordenamos por el nombre del exoplaneta. Después añadimos la columna 'rowid' que sirve como índice y reseteamos índice implícito para que quede ordenado

```python
PSCP_api = pd.DataFrame(exoplanets_json)

PSCP_api = PSCP_api.sort_values('pl_name')

PSCP_api['rowid'] = range(1, len(PSCP_api) + 1)
PSCP_api = PSCP_api[['rowid'] + PSCP_api.columns[:-1].tolist()]

PSCP_api = PSCP_api.reset_index(drop=True)
PSCP_api
```

Creamos archivo csv en local para no realizar la descarga cada vez (2-5 minutos de descarga)

```python
PSCP_api.to_csv('../data/pscomppars_API.csv', sep=',', index=False)
```
