# Limpieza de datos

Para nuestro análisis exploratorio necesitaremos los datos de las siguientes columnas (* datos imprescindibles):

<u>Planeta</u>
- pl_name  
- (*)pl_rade: radio en radios terrestres (R<sub>Terrestre</sub>)
- (*)pl_bmasse: masa en masas terrestres
- (*)pl_eqt: temperatura de equilibrio
- pl_orbper: periodo orbital (días)
- pl_orbsmax: semieje mayor (distancia a la estrella)
- (*)pl_insol: insolación recibida

<u>Estrella</u>  
- hostname: nombre de la estrella
- st_teff: temperatura efectiva
- st_rad: radio
- (*)st_mass: masa
- st_met: metalicidad
- st_spectype: tipo de estrella/tipo espectral (G, K, M…)  
>>Principales Tipos Espectrales (de más caliente a más frío):  
>>>O (Azules): (>30 000 K) Muy masivas, raras y brillantes.  
>>>B (Azules/Blancas): (10 000 - 30 000 K) Brillantes, helio ionizado.  
>>>A (Blancas): (7 500 - 10 000 K) Líneas de hidrógeno fuertes.  
>>>F (Blanco-Amarillas): (6 000 - 7 500 K)   
>>>G (Amarillas): (5 200 - 6 000 K) Estrellas como el Sol.  
>>>K (Naranjas): (3 500 - 5 200 K) Más frías y tenues que el Sol.  
>>>M (Rojas): (2 500 - 3 500 K) Las más abundantes, muy frías.

>>Clases de Luminosidad (Tamaño):  
>>>I: Supergigantes  
>>>II: Gigantes luminosas  
>>>III: Gigantes  
>>>IV: Subgigantes  
>>>V: Secuencia Principal (enanas)  

- st_age: edad estimada
- st_rotp: periodo de rotación

<u>Sistema</u>  
- (*)sy_dist: distancia en parsecs
- sy_pnum: número de planetas en el sistema
- (*)discoverymethod: método de detección
- disc_year: año descubierto

El año de descubrimiento tiene el tipo float porque hay un planeta cuyo año desconocemos. Dado que no es un dato esencial para nuestro análisis, buscamos cuál es el año más común para el método de descubrimiento para darle ese valor.

```python
exoplanets[exoplanets['disc_year'].isna()]
```

```python
# Máscara de años que faltan
falta = exoplanets['disc_year'].isna()

# Método de descubrimiento del planeta al que le falta 
metodo = exoplanets['discoverymethod'][falta].iloc[0]

# Sustituimos el valor
exoplanets.loc[falta, 'disc_year'] = exoplanets.loc[exoplanets['discoverymethod'] == metodo, 'disc_year'].mode().iloc[0]

# Cambiamos a tipo int el año de descubrimiento
exoplanets['disc_year'] = exoplanets['disc_year'].astype(int)
```

Comprobamos que todos los años están informados y son de tipo int.

```python
exoplanets.info()
```

Para nuestro análisis es fundamental que dispongamos de una serie de datos con los datos imprescindibles que mencionamos al comienzo

```python
# Filtrado base: quedarnos con filas que tengan al menos radio, masa y distancia
exoplanets = exoplanets.dropna(subset=['pl_rade','pl_bmasse','sy_dist'], how='any')
```

Una vez limpios los datos nos quedamos con 6053 planetas.

No todos los planetas tienen semieje mayor o radio orbital promedio 'pl_orbsmax' informado. Vamos a calcularlo usando la tercera ley de Kepler.

```python
# 1. Calcular periodo en años
exoplanets['P_yr'] = exoplanets['pl_orbper'] / 365.25

# 2. Calcular semieje mayor con Kepler SOLO donde falte pl_orbsmax
faltan = exoplanets['pl_orbsmax'].isna() & exoplanets['P_yr'].notna() & exoplanets['st_mass'].notna()

exoplanets.loc[faltan, 'pl_orbsmax'] = (
    (exoplanets.loc[faltan, 'P_yr']**2 * exoplanets.loc[faltan, 'st_mass'])**(1/3)
)

# 3. Borrar columna auxiliar
exoplanets.drop(columns=['P_yr'], inplace=True)
```

Comprobamos que la mayoría se ha informado.
